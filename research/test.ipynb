```python
collection_name = "financebot"
```

```python
vstore = AstraDBVectorStore(
    embedding= embedding,
    collection_name= collection_name,
    api_endpoint=ASTRA_DB_API_ENDPOINT,
    token=ASTRA_DB_APPLICATION_TOKEN,
    namespace=ASTRA_DB_KEYSPACE
    )
```

```python
from langchain_community.document_loaders import PyPDFLoader
```

```python
loader = PyPDFLoader("/Users/shehbazpatel/Documents/Ineuron_Projects/Generative_AI/Finance-Bot/data/finance_data.pdf")
```

```python
pages = loader.load()
```

```python
len(pages)
```

```python
pages = pages[10:20]
```

```python
pages[0].page_content
```

```python
for i, doc in enumerate(pages):
    print(i, doc)
```

```python
raw_text = ''
for i, doc in enumerate(pages):
    text = doc.page_content
    if text:
        raw_text += text
```

```python
print(raw_text)
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size = 500,
    chunk_overlap = 100,
)
```

```python
texts = text_splitter.split_text(raw_text)
```

```python
len(texts)
```

```python
texts[0]
```

```python
from langchain_core.documents import Document
```

```python
docs = []
for i in range(len(texts)):
    doc = Document(page_content=texts[i])
    docs.append(doc)
```

```python
docs
```

```python
vstore.add_documents(docs)
```

```python
retriever = vstore.as_retriever(search_kwargs = {"k":3})
```

```python
retriever.get_relevant_documents("what is Market For Registrant’s Common Equity?")
```

```python
PRODUCT_BOT_TEMPLATE = """
    Your finance bot is an expert in finance related advice.
    Ensure your answers are relevant to the query context and refrain from straying off-topic.
    Your responses should be concise and informative.

    CONTEXT:
    {context}

    QUESTION: {question}

    YOUR ANSWER:
    
    """
```

```python
from langchain_core.prompts import ChatPromptTemplate
```

```python
prompt = ChatPromptTemplate.from_template(PRODUCT_BOT_TEMPLATE)
```

```python

from langchain_openai import ChatOpenAI
```

```python
llm = ChatOpenAI()
```

```python
from langchain_core.runnables import RunnablePassthrough
```

```python
from langchain_core.output_parsers import StrOutputParser
```

```python
chain = (
        {"context": retriever, "question": RunnablePassthrough()}
        | prompt
        | llm
        | StrOutputParser()
    )
```

```python
chain.invoke("what is Market For Registrant’s Common Equity?")
```
